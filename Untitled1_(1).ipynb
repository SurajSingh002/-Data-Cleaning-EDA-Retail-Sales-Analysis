```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("retail_sales_dataset.csv")
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df.columns
```

```python
df['age'] = df['age'].fillna(df['age'].median())
df['total_amount'] = df['total_amount'].fillna(df['total_amount'].mean())
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
Q1 = df['total_amount'].quantile(0.25)
Q3 = df['total_amount'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
outliers = df[(df['total_amount'] < lower) | (df['total_amount'] > upper)]
print("Outliers:", len(outliers))
```

```python
df = df[(df['total_amount'] >= lower) & (df['total_amount'] <= upper)]
```

```python
sns.histplot(df['total_amount'], kde=True)
plt.title("Sales Distribution")
plt.show()
```

```python
sns.barplot(x='product_category', y='total_amount', data=df)
plt.xticks(rotation=45)
plt.title("Sales by Category")
plt.show()
```

```python
df.groupby('date')['total_amount'].sum().plot()
plt.title("Sales Trend")
plt.show()
```

```python
sns.scatterplot(x='age', y='total_amount', data=df)
plt.title("Age vs Spending")
plt.show()
```

```python
sns.barplot(x='gender', y='total_amount', data=df)
plt.title("Sales by Gender")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation")
plt.show()
```

```python
df.to_csv("cleaned_retail_data.csv", index=False)
```

```python
Missing Values:
 Age → filled using median
 Total Amount → filled using mean

Duplicates:
 Removed duplicate rows

Outliers:
 Removed using IQR method

Transformations:
 Column names standardized
 Date converted to datetime
```
